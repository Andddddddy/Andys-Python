# 成績計算器

```python
import tkinter as tk

window = tk.Tk()
window.title("Grade Calculator")
window.geometry("600x400")
window.configure(background = "white")

header_label = tk.Label(window, text = "成績計算器", font = ("Helvetica", 20), width = 15, height = 2)
header_label.pack(side = tk.TOP)

def cal_g():
    w_g1 = float(g1_entry1.get())*float(g1_entry2.get())
    w_g2 = float(g2_entry1.get())*float(g2_entry2.get())
    w_g3 = float(g3_entry1.get())*float(g3_entry2.get())
    w_g4 = float(g4_entry1.get())*float(g4_entry2.get())
    w_g5 = float(g5_entry1.get())*float(g5_entry2.get())
    w_g = round((w_g1 + w_g2 + w_g3 + w_g4 + w_g5) / (float(g1_entry2.get())+float(g2_entry2.get())+float(g3_entry2.get())+float(g4_entry2.get())+float(g5_entry2.get())), 2)
    result = "總成績為{}".format(w_g)
    # 將計算結果更新到 result_label 文字內容
    result_label.configure(text = result)
    

#第一列
g1_frame = tk.Frame(window)
g1_frame.pack(side = tk.TOP)

g1_label1 = tk.Label(g1_frame, text = "課程成績(分)：", font = ("Helvetica", 12), width = 15, height = 2)
g1_label1.pack(side = tk.LEFT)

entry1_1 = tk.DoubleVar()
g1_entry1 = tk.Entry(g1_frame, textvariable = entry1_1)
g1_entry1.pack(side = tk.LEFT)

g1_label2 = tk.Label(g1_frame, text = "學分數：", font = ("Helvetica", 12), width = 15, height = 2)
g1_label2.pack(side = tk.LEFT)

entry1_2 = tk.DoubleVar()
g1_entry2 = tk.Entry(g1_frame, textvariable = entry1_2)
g1_entry2.pack(side = tk.LEFT)

#第二列
g2_frame = tk.Frame(window)
g2_frame.pack(side = tk.TOP)

g2_label1 = tk.Label(g2_frame, text = "課程成績(分)：", font = ("Helvetica", 12), width = 15, height = 2)
g2_label1.pack(side = tk.LEFT)

entry2_1 = tk.DoubleVar()
g2_entry1 = tk.Entry(g2_frame, textvariable = entry2_1)
g2_entry1.pack(side = tk.LEFT)

g2_label2 = tk.Label(g2_frame, text = "學分數：", font = ("Helvetica", 12), width = 15, height = 2)
g2_label2.pack(side = tk.LEFT)

entry2_2 = tk.DoubleVar()
g2_entry2 = tk.Entry(g2_frame, textvariable = entry2_2)
g2_entry2.pack(side = tk.LEFT)

#第三列
g3_frame = tk.Frame(window)
g3_frame.pack(side = tk.TOP)

g3_label1 = tk.Label(g3_frame, text = "課程成績(分)：", font = ("Helvetica", 12), width = 15, height = 2)
g3_label1.pack(side = tk.LEFT)

entry3_1 = tk.DoubleVar()
g3_entry1 = tk.Entry(g3_frame, textvariable = entry3_1)
g3_entry1.pack(side = tk.LEFT)

g3_label2 = tk.Label(g3_frame, text = "學分數：", font = ("Helvetica", 12), width = 15, height = 2)
g3_label2.pack(side = tk.LEFT)

entry3_2 = tk.DoubleVar()
g3_entry2 = tk.Entry(g3_frame, textvariable = entry3_2)
g3_entry2.pack(side = tk.LEFT)

#第四列
g4_frame = tk.Frame(window)
g4_frame.pack(side = tk.TOP)

g4_label1 = tk.Label(g4_frame, text = "課程成績(分)：", font = ("Helvetica", 12), width = 15, height = 2)
g4_label1.pack(side = tk.LEFT)

entry4_1 = tk.DoubleVar()
g4_entry1 = tk.Entry(g4_frame, textvariable = entry4_1)
g4_entry1.pack(side = tk.LEFT)

g4_label2 = tk.Label(g4_frame, text = "學分數：", font = ("Helvetica", 12), width = 15, height = 2)
g4_label2.pack(side = tk.LEFT)

entry4_2 = tk.DoubleVar()
g4_entry2 = tk.Entry(g4_frame, textvariable = entry4_2)
g4_entry2.pack(side = tk.LEFT)

#第五列
g5_frame = tk.Frame(window)
g5_frame.pack(side = tk.TOP)

g5_label1 = tk.Label(g5_frame, text = "課程成績(分)：", font = ("Helvetica", 12), width = 15, height = 2)
g5_label1.pack(side = tk.LEFT)

entry5_1 = tk.DoubleVar()
g5_entry1 = tk.Entry(g5_frame, textvariable = entry5_1)
g5_entry1.pack(side = tk.LEFT)

g5_label2 = tk.Label(g5_frame, text = "學分數：", font = ("Helvetica", 12), width = 15, height = 2)
g5_label2.pack(side = tk.LEFT)

entry5_2 = tk.DoubleVar()
g5_entry2 = tk.Entry(g5_frame, textvariable = entry5_2)
g5_entry2.pack(side = tk.LEFT)


#Result
result_label = tk.Label(window)
result_label.pack()


#Button
b = tk.Button(window, text = "開始計算",font = ("Helvetica", 12), width = 15, height = 2, command = cal_g)
b.pack(side = tk.TOP)


window.mainloop()
```
