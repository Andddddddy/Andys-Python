# 停車費


```python
t = input().split()

h1, m1 = int(t[0].split(":")[0]), int(t[0].split(":")[1])
h2, m2 = int(t[1].split(":")[0]), int(t[1].split(":")[1])

if h2-h1 == 0:
    if m2-m1 <= 30:
        print(0)
    else:
        print(30)

elif h2-h1 == 1:
    if m1 >= m2:
        if (60-m2+m1)/30 <= 1:
            print(0)
        else:
            print(30)
            
    else:
        print(60)

elif h2-h1 == 2:
    if m1 >= m2:
        print(60)
            
    else:
        print(120)

else:
    print(120)
#done
```
