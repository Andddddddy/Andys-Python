# 及格名單


```python
m = input().split(",")
e = input().split(",")

mL = list(set(m) - set(e))
mL.sort()
mL1 = ",".join(mL)  #將資料型態變回「字串」，就能print出想要的答案
print(mL1)

eL = list(set(e) - set(m))
eL.sort()
eL1 = ",".join(eL)
print(eL1)

meL = list(set(m) & set(e))
meL.sort()
meL1 = ",".join(meL)
print(meL1)

print(len(set(m) | set(e)))
#done
```
