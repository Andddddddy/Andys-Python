# 兩個字典的鍵值配對


```python
# 先建立輸入資料數量的input

num = int(input())
```

```python
# 根據資料數目用for跑出對應數量的input
# 接著建立：{} -> {"P":{}, "Q":{}} -> {"P":{"key0": "2"}, "Q":{"key2": "3"}}

catedict = {}
for row in range(num):
    rowList = input().split()  # 根據資料數目跑出input數量，並且分隔元素。
    if not rowList[-1] in catedict:  # 加上條件式，當catedict裡面沒有該rowList名稱，才會執行。
        catedict[rowList[-1]] = {}
    catedict[rowList[-1]][rowList[0]] = rowList[1]
```

```python
k = list(catedict.keys())
k.sort()
k[1]
```

```python
resultSet = set(catedict[k[0]].items()) & set(catedict[k[1]].items())
resultSet
```

```python
resultList = list(resultSet)
resultList.sort()
resultList
```

```python
for key, value in resultList:
    print("{}: {} is present in both {} and {}".format(key, value, k[0], k[1]))
```

```python
num = int(input())

catedict = {}
for row in range(num):
    rowList = input().split()
    if not rowList[-1] in catedict:
        catedict[rowList[-1]] = {}
    catedict[rowList[-1]][rowList[0]] = rowList[1]
    
k = list(catedict.keys())
k.sort()
resultSet = set(catedict[k[0]].items()) & set(catedict[k[1]].items())

resultList = list(resultSet)
resultList.sort()

for key, value in resultList:
    print("{}: {} is present in both {} and {}".format(key, value, k[0], k[1]))
```
