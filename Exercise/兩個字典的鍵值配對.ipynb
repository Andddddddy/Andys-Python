```python
# input

num = int(input())
```

```python
# forinput
# {} -> {"P":{}, "Q":{}} -> {"P":{"key0": "2"}, "Q":{"key2": "3"}}

catedict = {}
for row in range(num):
    rowList = input().split()  # input
    if not rowList[-1] in catedict:  # catedictrowList
        catedict[rowList[-1]] = {}
    catedict[rowList[-1]][rowList[0]] = rowList[1]
```

```python
num = int(input())

catedict = {}
for row in range(num):
    rowList = input().split()
    if not rowList[-1] in catedict:
        catedict[rowList[-1]] = {}
    catedict[rowList[-1]][rowList[0]] = rowList[1]
    
k = list(catedict.keys())
k.sort()
resultSet = set(catedict[k[0]].items()) & set(catedict[k[1]].items())

resultList = list(resultSet)
resultList.sort()

for key, value in resultList:
    print("{}: {} is present in both {} and {}".format(key, value, k[0], k[1]))
```
