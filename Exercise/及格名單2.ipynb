# 及格名單2


```python
dic = {"name":0, "Chinese":1, "English":2, "Math":3, "Science":4, "Society":5}
n = int(input())

dataList = []
for num in range(n):
    grades = input().split()
    dataList.append(grades)
    
choice = input().split()

result = []
for i in range(5):
    if int(dataList[i][dic[choice[0]]]) >= 60 or int(dataList[i][dic[choice[1]]]) >= 60:
        result.append(dataList[i][0])
        result.sort()
        
if result == []:
    print("NO")
else:
    print(",".join(result))
```
